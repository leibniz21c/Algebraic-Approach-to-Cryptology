# CH02. 다표식 치환 암호
---
<strong>2.1. 동음이의 암호</strong><br>
챕터1에서 한 암호방식은 단표식 치환 암호들이고, 이는 동일한 치환 규칙이 유지가된다.<br>
반면에 다표식 암호는 한 번에 한 문자를 치환하지만 문자마다 다른 치환 규칙을 사용한다.<br>
예를들어 한 문자를 여러 문자로 암호화하는 암호를 가리켜 Homophonic cipher(동음이의 암호)라 한다. 일반적으로 동음이의어라는 말은 서로 철자가 다른 두 단어가 발음이 같은 관계를 의미한다. 암호학에서 동음이의어란 암호문에서는 다른 형태를 갖지만 동일한 문자로 복호화되는 문자를 의미한다. 굉장히 간단하게도 이 방법을 사용하면 문자 빈도 분석에 대하여 일차적으로 안전하다. <br><br>

>케르크호프스 원리 : Key를 제외한 모든 정보가 알려지더라도 암호체계는 안전해야한다.

만약, 케르크호프스 원리를 기본 가정으로 삼는다면, 암호의 원리를 추정하기 위해 암호체계를 훔칠 필요가 없을 때도 있다. 만약 암호문을 해독하는 입장에서 동음이의 암호가 사용된다는 것을 어떻게 추정할 수 있을까? 일반적으로 동음이의 암호체계에는 문자가 26개 이상이다. 그러나 모든 암호문 문자가 실제로 메시지에 쓰이지 않을 수도 있고 여러가지 다른 변수가 많다. 실제 예시에서 추정을 해보자.<br><br>

첫 번째 단계는 암호문 문자 각각에 대한 출현 빈도를 표로 만들었을때 다음과 같은 상황이 발생했다고 생각하자.<br>
> 암호문에 나타난 문자는 모두 23개이다. 

이는 다루고 있는 언어의 전체 문자 수가 26개 이하이거나, 모든 문자를 다 사용하지 않아도 되는 다중문자 치환 방식을 사용했거나, 아니면 평문에서 일부 문자가 나타나지 않았다는 의미이다.<br><br>

두 번째 단계는 조사한 암호문의 관찰 빈도와 영어 텍스트의 기대 빈도를 비교하는 것이다. 다음과 같은 상황이 발생했다고 생각하자.<br>

표를 보면 주어진 암호문은 기대 빈도가 가장 낮은 문자들이 우연히 사용되지 않은 단순 치환 암호라고 추정할 수 충분히 있다. 만약 동음이의 암호가 사용되었다면 기대 빈도가 낮은 문자는 더 많이 나타나고 기대 빈도가 높은 문자는 오히려 더 적게 나타났을 것이다. 즉, 알파벳 발생 빈도의 분포형태가 일반적인 문자 빈도 분포와 많이 달라질 것이다. 이런 관찰 결과를 보다 수학적으로 분석하기 위한 도구가 Indexx of Coincidence(동시발생지수)이다. <br><br>

<strong>2.2. 문자 빈도와 동시발생지수</strong><br>
동시발생지수를 공부하기전에 확률이라는 토픽을 조금만 살펴보자. 확률이란 어떤 사건이 일어나는 경우의 수가 K가지 일 때, 어떤 사건이 일어나는 확률이 
$ \frac {K}{N}$인 것을 뜻한다. 그래서 확률 $p$는 다음과 같은 범위를 가진다. $ 0 \le p \le 1 $ <br><br>

먼저 각 문자가 동일한 빈도로 출현한다는 가정에서 무작위로 분포되어 있다고 가정할때, 첫 번째 뽑은 문자가 a일 확률은 $\frac{1}{26}$이다. 두 번째 선택에서 a를 뽑을 확률이 $\frac{1}{26}$이므로, 첫 번째에서 뽑은 문자와 두 번째에서 뽑은 문자가 연속으로 a일 확률은 $\frac{1}{26} \times \frac{1}{26}$이다. 각 문자별로 동시에 발생했다고 가정할때 확률의 합은 다음과 같다.
$$ \frac{1}{26^2} + \frac{1}{26^2} + \cdots \frac{1}{26^2} = \sum_{k=1}^{26} \frac{1}{26^2} = \frac{1}{26} \approx .038 $$
이처럼 선택된 텍스트에서 문자를 하나씩 두 번 뽑을 때, 서로 같은 문자를 뽑을 확률을 동시발생지수라 한다. 무작위의 영어 텍스트(일반적으로 문자 빈도 분석에서 사용하는 것과 다른 분포)에서 동시발생지수는 대략 0.038, 즉 3.8%이다.<br><br>

실제 영어 텍스트(문자 빈도 분석에 사용하는 분포)에서 문자를 뽑는다고 가정하자. 일단 a가 뽑힐 확률은 8.2%, 소수로 나타내면 0.082이다. 이때 a를 연속으러 뽑을 확률은 $(0.082)^2$ 이다. 모든 문자에 대해서 합을 구하면 다음과 같다.
$$ (.082)^2 + (.015)^2 + (.028)^2 + \cdots + (.001)^2 \approx .066 $$
즉, 실제 영어 텍스트의 동시발생지수는 대략 0.066, 즉 6.6%이다.<br><br>

무작위 영어 알파벳의 분포는 0.038에 가까운 값이 나올 것이고, 의미가 있는 텍스트에서는 당연히 0.066에 가까운 값이 나올 것이다.<br><br>

첫 번째로 텍스트에 단순 치환 암호를 적용하더라도 동시발생지수는 변하지 않는다. 더하는 순서는 바뀌더라도 총합은 같기 때문이다. 그러므로 단순 치환 암호를 사용한 암호문의 동시발생지수는 대략 0.066이다. 하지만 동음이의 암호를 사용해 암호문을 만들었다면 동시발생지수는 분명히 달라진다. 동음이의 방식에서는 문자 간의 출현 빈도 차가 적어지기 때문에 동시발생지수는 동시발생지수는 0.038과 0.066 사이가 된다. <br><br>

위의 예시에서 다루던 예시를 계산하면 0.068이 나오는데 단순 치환 암호를 사용했다는 쪽으로 무게가 실린다. 수업에서 이 예시에 대해서 조금더 다루겠다.<br><br>

동음이의암호는 여기까지 하자. 이제 르네상스 시대로 넘어올때이다.

<strong>2.3. 알베르티 암호</strong><br>
1466년 혹은 1467년 초에 'De Componendis Cifris(암호에 대한 논고)'라는 25쪽짜리 논문이 유럽에 발표된다. 이 논문은 유럽의 초기 암호기술과 암호해독에 대한 학술저작물로서, 유럽 최초로 문자 빈도 분석법을 사용하는 방법과 무효 문자, 동음이의암호의 개념에 대해 설명한다. 게다가 최초의 치환 암호기계이면서 최초의 진정한 다표식 암호방식(암호문이 진행중에 치환 방식이 바뀐다는 순수성에 근거, 동음이의암호는 치환 방식이 바뀌는 것에 의의를 두기 힘들다.)이라 할 수 있는 암호 원판에 대한 소개가 실려있다. 이 논문의 저자는 '레온 바티스타 알베르티'라는 사람인데, 이는 이탈리아의 저술가이자 미술가이며 건축가, 운동선수, 철학지인 르네상스 시대의 팔방미인이라고 소개되는 것 같다. 일단 암호 원판을 보자.<br>

위의 그림을 보면 일단은 평문(소문자), 암호문(대문자)에 대해서 순서가 있게 둘다 배치가 되어있음을 알 수 있다. 알베르티 암호 원판은 바깥 쪽의 큰 판, 안 쪽의 작은 판으로 이루어져 있으며, 이 것들은 분리가 되어있고 중심에 핀을 꽂아 회전할 수 있게 되어 있다. 위의 사진은 알파벳이 순서에 맞게 배치되어 있으며, 이를 돌려서 간단하게 덧셈암호에 대해서 단표식 단일 치환을 할 수 있음을 알 수 있다. 하지만, 강의화면에서 볼 수 있는 원판에서는 암호문부분의 알파벳 순서는 무질서하게 분산시켜져 있음을 알 수 있다. 이때 작은 원판을 한 자리에 고정하고 문자를 대응시키면 고전적이지만 일반적인(무질서한) 단표식 단일 치환 암호가 된다.<br><br>

실습을 감상하기전에 2가지 정의를 하자. 첫 번째는 암호문의 문자 중에서 하나의 문자를 선택해 '지표문자'라고 부르자. 원판을 돌릴때 기준이 되는 문자라고 생각해주면 편할 것 같다. 두 번째는 평문 문자중에서 하나의 문자를 선택해 '키 문자'라고 부르자. 그리고 이를 암호문의 시작 부분에 적기로 약속하자. 즉, 키 문자가 지표 문자와 동일한 위치에 오도록 원판을 회전하여 암호표처럼 활용할 수 있다. 첫 번째 실습을 보자.<br><br>

실습에서도 알 수 있듯이, 이는 <strong>단표식</strong> 단일치환 암호방식이다. 즉, 문자 빈도 분석법에 굉장히 취약한 암호방식이다. 하지만, 알베르티는 여기서 좋은 아이디어를 제시하여 문자 빈도 분석법에 대비할 수 있는 최초(엄밀하게는 아니지만)의 다표식 치환 암호방식을 고안해낸다. 과정은 간단하다. 단어 3~4개를 암호화하고 나면, 원판을 새로 돌려 암호 규칙을 바꿔주고, 암호문에 새로운 키 문자를 삽입해 암호 변경을 표시하는 방법이다. 두 번째 실습을 보자.<br><br>

생각보다 쉬운 방식으로 다표식 암호화 방식을 구현할 수 있음을 알았다. 그럼 이 알베르티 암호의 복호화는 어떻게 진행될까? 우선 복호화를 하는 입장과 부호화를 하는 입장 모두 같은 암호 원판을 가지고 있어야 한다. 같은 암호 원판이라는 말은, 애초에 안쪽판이 무질서하게 분산되어져 있는데, 이러한 단일 치환 규칙의 순서는 맞아야 한다는 뜻이다. 추가적으로 부호, 복호화하는 사람은 지표 문자를 서로 공유하고 있어야한다. 만약, 이 두 조건이 충족되면, 암호문에 키 문자가 나올 때마다 지표 문자와 동일한 위치에 오도록 원판을 돌리고, 알파벳끼리 대응시켜 복호화를 진행하면된다. 이 까지 읽어보면 이번 주 실습은 어렵지 않을까라는 생각이 들 수 있는데, 생각보다 쉬우니 걱정말고 진행하자. 복호화 실습을 보자.<br><br>

제일 재미있는 부분이다. 알베르티 암호를 부호화, 복호화에 대해서 공부를 했는데, 이제 남은 것은 어떻게 공격할까에 대한 문제이다. 잘 생각해보자. 만약 암호체계가 알베르티 암호라는 사실을 알고, 암호문에 등장하는 소문자가 키 문자라는 사실을 알더라도 원판을 모른다면 복호화할 수 없다. 만약 키 문자를 충분히 자주 바꾸고, 메시지가 그다지 길지 않아 키문자가 중복 사용되지 않는다면, 빈도 분석 공격을 사용하기도 힘들다. 키문자가 중복 사용이 된다면, 메시지 길이가 (충분할지는 모르겠지만)길어지기에, 특정 키문자가 사용된 파트만 분할하여 빈도 분석 공격을 진행하는 것은 가능은 하다. 그래서 메시지의 길이가 충분히 길어서 원판의 위치를 중복해서 사용하는 경우가 많아지면(빈도 분석 공격을 하기에 충분히) 빈도 분석 공격을 통해 알베르티 암호를 해독할 가능성은 충분히 있다. 추가적으로 중복해서 사용하는 키 문자가 많아진다면 암호판 자체를 만들어버리는 것도 충분히 가능하다. 실습을 보고 지나가자.



<strong>2.4. 정사각형 암호표 : 트리테미우스 암호표</strong><br>

한 칸, 두 칸, 세 칸 등 처음 상태로 되돌아올 때까지 계속 회전하면서 위의 암호표가 완성된다. 만약 평문 "abc"를 암호문으로 바꾸면 "BDF"가 된다. 암호 원판과 유사한 정보를 가지고 있으나, 우선적으로 모든 정보를 한꺼번에 보여주고 있음을 알 수 있다. 부호화를 진행하는 사람이 원할때마다 키 문자를 쓰고 원판의 위치를 바꾸는 형태가 아니라, 정사각형 암호표를 따라 차례대로 내려가면서 문자 하나마다 암호문 문자로 하나씩 치환하는 것이다. 암효표의 맨 마지막 줄에 이르면 처음부터 다시 시작한다. <br><br>

부호화를 하는 사람과 복호화를 하는 사람이 단어나 구로 된 키를 미리 약속을 한다 이를 Keyword(키워드)라고 하자. 그리고 키워드를 필요한 만큼 반복해서 평문의 위 칸에 적는다. 이후 평문의 각 문자에 대해 해당 키와 대응되는 알파벳 행에서 암호문 문자를 찾는다. 실습을 보며 확인하자.<br><br>

힐 암호의 경우(단표식 다중 문자 치환)는 평문에서는 같은 문자이더라도 위치에 따라 암호문에서는 다른 문자로 나타날 수 있는데, 어떤 경우에는 3개의 s가 각각 M, E, X로 치환된다. 이런 형태의 다표식 암호를 Repeating-Key Cipher(반복키 암호)라고 부른다. 반복키 암호의 키는 반드시 그런 것은 아니지만 단어나 어구가 가장 흔하게 사용된다. 이렇게 벨라조는 순서가 다양한 키 알파벳과 암호문 알파벳을 사용해 더 복잡한 암호체계를 만들었다. <br><br>

그런데 대수적 관점에서 보면 트리테미우스 암호표를 사용하는 것이 모듈러 연산에 있어서 표현이 간결하다는 장점이 있다. 식은 다음과 같다.
$$ C_i \equiv P_i + k_{i \,\bmod \,\ length \,\ of \,\ keyword} \,\bmod \,\ 26 $$
식에 대한 설명은 수업시간에 진행하겠다. 언듯보면 덧셈암호와 비슷한 이러한 방식은 현대 디지털 스트림 암호의 초기 형태이다.<br><br>

 위의 식을 통해서 간단하게 부호화할 수 있다. 추가적으로 복호화도 식을 조금만 변형하면 쉽게 가능하다. <br><br>
