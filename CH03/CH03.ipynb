<strong>3.0. Introduction of Transpose Cipher</strong><br>
지금까지 살펴본 암호는 모두 한 문자가 다른 문자로 바뀌거나 한 문자열이 다른 문자열로 대체되는 치환 방식이었다. 그에 반해서, 전치 암호는 문자를 치환하는 것이 아니라 단지 문자의 자리를 이동하는 방식이다.

<strong>3.1. 스키테일 암호</strong><br>
전치 암호의 역사는 최소 고대 그리스로 거슬러 올라가야한다. 기록에 남아 있는 최초의 전치 암호는 사이테일(Scytale)이다. 단어의 음역은 '스키테일'이지만 대개 c를 발음하지 않는 영어식 발음을 써서 '사이테일'이라고 많이 부른단다. 한국에서는 '스키테일'이라고 부르는 것 같으니 일단 '스키테일'이라고 하자.<br><br>

스키테일 암호는 고대 그리스의 폴리스중 하나인 스파르타에서 처음 사용했을 것으로 추정된다. 스파르타의 라이산더(Lysander)장군이 사용했다는 기록이 남아있다. 스키테일은 막대기를 의미한다. 라이산더 장군이 나무막대를 암호 장치로 사용한 지 수세기가 지나서야 처음으로 로마의 역사학자 플루타크코스가 스키테일 암호에 대해 기록했다. 원문은 다음과 같다. (역사적인 부분을 사실 빼려고 했는데, 읽어볼만큼 재미있으니까 짚고 넘어가겠습니다.)

> 스파르타인이 사용한 통신문은 다음과 같은 성질을 띤다. 스파르타의 시의회나 민선 장관이 장군을 파견할 때, 길이와 두께가 정확히 같은 둥근 나무막대 두 개를 만들어 하나는 보관하고 다른 하나는 파견 장군에게 준다. 이 나무막대를 '스키테일'이라 하다. 비밀리에 중요한 메시지를 보내고자 할 때마다 가죽끈처럼 양피지로 길고 가는 띠를 만들어 빈 공간이 생기지 않도록 나무막대에 감는다. 그다음, 감겨 있는 양피지 위에 원하는 내용을 적는 것이다. 메시지를 다 완성하면 양피지 띠를 풀어서 문서만 보낸다. 띠 모양의 문서에는 아무 연관이 없는 문자들이 무질서하게 나열되어 있어서 그것을 받은 사령관은 의미를 파악하지 못한다. 그러나 가지고 있던 사이테일에 문서를 감으면 원래의 나선형 모양이 완벽하게 복구되어 나중 문자와 앞 문자와 정확하게 연결되고 나무막대를 따라 글을 읽으면 메시지 내용이 나온다. 나무막대와 마찬가지로 양피지 띠도 스키테일이라 부른다.

충분히 상상이 되지만, 시각적으로 확인해보자. [스키테일 암호 구현 유튜브 영상](https://www.youtube.com/watch?v=Iwf4rZXVXjw)<br><br>

이제 이것을 분석해보자. 만약 둘레를 따라 정확히 철자 세 개를 적을 수 있고 띠를 11번 감을 수 있는 길이의 나무막대가 있다면, 철자를 써넣을 수 있는 공간이 $ 3 \times 11 $인 격자판 또는 행렬이된다.<br><br>
$ \begin{bmatrix}
\rightarrow & g & o & t & e & l & l & t & h & e & s & p & \rightarrow \\
\rightarrow & a & r & t & a & n & s & t & h & o & u & w & \rightarrow \\
\rightarrow & h & o & p & a & s & s & e & s & t & b & y & \rightarrow \\
\end{bmatrix} $<br><br>

위와 같이 구성된 스키테일을 풀면, 아래와 같은 암호문이 완성된다.
> GAHOR OTTPE AALNS LSSTT EHHSE OTSUB PWYAZ

참고로 마지막의 AZ는 메시지의 길이를 숨기기 위해 무효 문자로 채웠다. 결국 스키테일을 풀었을 때, 격자판에서 읽는 순서는 다음과 같다.
$ \begin{bmatrix}
\downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow \\
 g & o & t & e & l & l & t & h & e & s & p \\
 a & r & t & a & n & s & t & h & o & u & w \\
 h & o & p & a & s & s & e & s & t & b & y \\
 \downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow \\
\end{bmatrix} $<br><br>

이처럼 문자를 격자판 모양으로 배치한 후 세로 방향으로 읽어 암호문을 만드는 방식을 대개 행렬형 전치(Columnar transposition)라 한다.<br>
자 그러면 사이테일 암호에 키가 있을까? 플루타크코스의 기록에 따르면 길이와 두께가 정확이 같은 둥근 나무막대 두 개가 필요하다. 실제로 반드시 일치해야 하는 것은 길이 보다는 두께라고 할 수 있다. 만약 한 번 감았을 때 4개의 문자가 감기는 나무막대를 사용해 암호문을 해독하려 한다면 종이 띠를 모두 감았을 때 다음 메시지를 얻는다.<br>

$ \begin{bmatrix}
\downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow \\
 g & r & p & l & s & e & e & u & y & \\
 a & o & e & n & s & h & o & b & a & \\
 h & t & a & s & t & h & t & p & z & \\
 o & t & a & l & t & s & s & w &   & \\
 \downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow & \downarrow  \\
\end{bmatrix} $<br><br>

행렬형 전치 방식으로 만든 암호문은 세로 방향으로 적어 가로 방향으로 읽으면 복호화된다. 하지만 이렇게 얻어진 메시지는 가로방향으로 읽어도 의미 없는 문자들의 나열이다. 물론 마지막 열은 공백이 있기 때문에 이것이 무효 문자라는 것도 알 수 있다. 따라서, 스키테일 암호에서는 나무막대의 둘레, 즉 <strong>격자의 행의 개수가 키</strong>가 된다. 위의 예제에서는 키가 3이된다. <br><br>

단, 무효 문자를 사용하지 않으면 상당히 쉽게 키를 유추할 수 있다. 예를 들어서 위의 암호문은 길이가 33이 였다. 격자판의 구조에 따라서, 가능한 경우의 수는 $1 \times 33, \,\ 3 \times 11, \,\ 33 \times 1$밖에 없다. 첫 번째와 마지막 경우는 자명한 경우이기에, 키가 3 또는 11이라고 쉽게 유추할 수 있다.

<strong>3.3. 순열암호</strong><br>
위와 같이 기하학적인 구조와 관련없이 의존하지 않고 문자의 위치를 바꾸는 방법이 있다. 바로 순열을 사용하는 방식이다. 역사적인 부분에 대해서는 넘어가겠다. 만약 평문 ruby wine이 있다면, 순열 
$ \begin{pmatrix}
1 & 2 & 3 & 4 \\
2 & 4 & 3 & 1 \\
\end{pmatrix} $에 의해서 다음과 같이 암호화 될 수 있다.<br><br>

UYBR IENW<br><br>

네 문자로 구성된 암호문 문자군 각각 1번 자리는 평문의 2번 문자가 차지하고, 2번 자리는 평문의 4번 문자가 차지하고 있다. 3번 자리에는 평문의 3번 문자가, 4번 자리에는 평문의 1번 문자가 온다. 다음은 이븐 아드 두라이힘(ibn ad-Duraihim)이 라는 사람의 특수한 순열이다. 이를 통해서 부호화를 해보자.<br><br>
$ \begin{pmatrix}
1 & 2 \\
2 & 1 \\
\end{pmatrix} $<br>
평문 : ruby wine<br>
암호문 : URYB IWEN<br><br>

순열 암호의 복호화는 역순열을 이용해서 간단하게 할 수 있다. 리뷰에서 했지만, 역순열은 다음 조건을 만족하는 순열 $ \sigma^{-1} $이다.
$$ \sigma \sigma^{-1} = e $$
이는 순열의 1행과 2행의 자리를 바꾸고, 1행에 대해서 열의 순서를 맞추면 간단하게 구할 수 있다.<br>
이후 암호문에 대해서 역순열을 이용하여 부호화를 하듯이 연산을 진행하면 복호화가 끝난다. <br><br>

순열 암호의 키는 순열 그 자체이다. 위의 예시에서는 키가 순서대로 $ [2 \,\ 4 \,\ 3 \,\ 1], \,\ [2 \,\ 1] $이다.<br>
그런데, 문자열을 이용해서 순열의 키를 정의할 수 있다. 예를 들어서 키가 TALE이라면, 알파벳 순서대로 따진다면, <br>
T가 4번째, A가 1번째, L이 3번째, E가 2번째이므로, 키가 $[4 \,\ 1 \,\ 3 \,\ 2]$라고 할 수 있다. <br><br>

그러면 사이즈가 n인 순열암호에 대해서 가능한 키의 개수는 몇개일까?<br>
당연히 $n!$로 자명하다. 만약 길이가 12만 되어도 가능한 경우의 수는 $479,001,600$개이므로 무차별대입공격에 어느정도 안전하다고 할 수 있다.
