<strong>4.0. Introduction</strong><br>
Chapter04를 기점으로 이제 현대적인 암호체계에 대해서 공부를 해보겠다. 암호학이라는 분야는 무조건 컴퓨터와 밀접하지는 않는다. 적어도, 앨런 튜링이 태어나기전까지는 말이다. 앞에서 보아왔듯이, 시저암호, 스키테일암호, 트리테미우스 암호 등 많은 암호체계들이 사실은 굉장이 아날로그틱 한 방식으로 주고받아지고, 분석 되어왔다. 지금부터는 컴퓨터라는 강력한 계산기를 가지고 현대적인 암호체계에 대해서 공부해보자. 신나지만 조급하지말고 천천히 완벽하게 학습하자.

<strong>4.1. 현대 암호의 핵심 원리</strong><br>
현대 암호의 핵심 원리는 클로드 섀넌(Claude Shannon)이 새롭게 정의한 개념으로서, 통계를 바탕으로 만들어진 암호해독 기법을 우려해서 두가지를 제안했다. 먼저, 조금만 생각해보자. 우린 이미 알고있다. 첫 번째, 사실 피할 방법이 많지만, 그래도 Chapter03까지 우리에게 굉장히 강한 무기처럼 다가온 방법이다. LFA(Letter Frequency Analysis), 바로 문자 빈도 분석법이다. 또, 다른 것은 어떤 것이 있을까? 동시 발생 지수를 계산하는 것 또한 통계를 바탕으로 만들어진 암호해독 기법이다. 추가적으로 Known-Plain text Attack(알려진 평문 공격)과 같은 방법이 있겠다. 무차별 대입 공격도 엄밀히 따지면, 통계를 바탕으로 둔다. (가장 불확실성이 높은 경우를 가정하고 공격하는 방법이라는 관점으로, 학자별 견해가 다르다.) 여담인데 "섀넌"이라는분은 컴퓨터 쪽 공부하면 몰라서는 안되는 인물중 한 명이다. 시간나면 찾아보자. 재미있는 일화가 많다. 정보이론의 아버지라고 불리는 섀넌의 유명한 정의중 "섀넌 엔트로피"등이 등장한다.<br><br>

자 다시 본론으로 돌아와서, 제안한 두 가지는 다음과 같다. 확산(Diffusion)과 혼돈(Confusion).<br>
확산은 문자 빈도나 다이어그래프 빈도처럼 한 번에 단지 몇 문자만 살펴도 되는 평문의 통계적 구조가 암호문에서는 '확산'되어 길이가 긴 문자열을 살펴야 하는, 즉 넓은 범위에 걸쳐 퍼져 있는 통계 구조가 되어야 한다는 것이다. 말이 참 어렵다. 어쨋든 요점은 확산은 평문과 암호문과의 관계를 감추는 성질을 말한다. 예를 들어서, 평문 한 비트의 변화가 암호문의 모든 비트에 영향을 미친다면, 확산의 좋은 예가 될 수 있겠다. <br>
혼돈은 암호문에 대한 간단한 통계자료가 주어졌을 때 키를 찾는 것이 굉장히 복잡해야 한다는 원리이다. 즉, 키와 암호문과의 관계를 감추는 성질을 말한다. 예를 들어서 현대 블록암호는 혼돈을 위해 치환 기법을 사용한다.<br><br>



<strong>4.2. 혼합 함수</strong><br>
이전에 대수학 수업을 진행할 때, 함수에 관해서 엄밀하게 정의하고(당시 교안에서 집합론에 근거하였으니 걱정말자.) 다루어보았다. 수체계에 기반을 두는 함수는 독립변수나 종속변수가 기존의 수체계나 새롭게 정의되는 수체계에 맞게 구조화된다. 예를 들어서 함수값이 실수가 될 수도 있는데, 행렬, 순열, 복소수, 사원수 등 여러가지 정의된 수체계로 확장가능하다. 19세기 말 무렵부터 수학자들은 조금 더 넓은 시각에서 함수를 바라보기 시작했는데, 이때 함수는 어떤 형태의 대상을 '투입'해서 정해진 규칙에 따라 같은 형태 또는 다른 형태의 대상을 '산출'하는 시스템으로 생각하기시작했다. 이때 산출되는 것은, 수식이나 일련의 지시어, 표 심지어 그림도 가능하다. 이러한 관점으로 다음 함수를 보자.

$$ f(P) = \text{ "알파벳을 원형으로 배열한 상태에서 문자 P를 알파벳순으로 3자리 이동"} $$
수식으로도 나타낼 수 있지만, 위의 함수는 함수의 조건을 모두 만족한다. 그러므로 함수이다. (시저 암호에 대한 이야기를 굳이 말로 표현한 것이다.)<br>
섀넌은 확산과 혼돈을 제공하기 위해 암호에 혼합 함수(mixing function)를 사용할 것을 제안했다. 섀넌은 암호학에서 말하는 혼합 함수를 정확하게 정의하지 않았고, 할 수 없다고 인정했다고 한다. 대략적으로는 혼합 함수는 공간 내 응집된 영역을 상당히 균일하게 전체 공간으로 분산시키는 함수라고 할 수 있다. 이 말을 풀이해보자. 공간 내 응집된 영역을 평문의 한 비트로 보자. 이 것이 균일하게 분산된다면, 이는 혼돈의 조건에 부합한다고 할 수 있다. 즉, 평문과 암호문과의 관계를 균일하게 분산시킴으로서 결과적으로 키와 암호문과의 관계를 균일하게 분산시키는 성질이다. 추가적으로 한 비트가 전체 공간으로 분산된다면, 이는 확산의 조건에 부합한다고 할 수 있다. 즉, 하나의 비트가 암호문 전체에 영향을 미치고 있는 것이다. 이러한 성질위에서 암호학에서 말하는 혼합 함수가 정의될 수 있는데, 섀넌은 정확하게 정의할 수 없는 개념이라고 인정했다. <br><br>

간단하게 말하면 확산과 혼돈을 제공하기 위한 함수를 설계할 것 인데, 이때까지 우리가 공부한 내용이 다나온다. 다음 함수를 보자.

$$ F(P_1, P_2, \cdots, P_n) = H(S(H(S(H(T(P_1, P_2, \cdots, P_n)))))) $$

이때, T는 문자 n개로 된 문자군에 작용하는 전치 암호이고, H는 문자 n개로 된 블록에 작용하는 너무 복잡하지는 않은 힐 암호이다. S는 블록의 각 문자에 작용하는 단순 치환 암호를 가리킨다. 개별적 단계에서는 공격이 간단하지만, 이런식으로 작정하고 결합하거나 중복하면 공격하기 굉장히 까다로워진다. 즉, 좋은 혼합 함수라고 볼 수 있다.

위의 그림을 보면 함수 F가 내부적으로 결합, 중복을 통해서 혼합 구조를 가지고 있는 함수임을 알 수 있다. 섀넌이 생각하는 혼합함수의 개념은 우선적으로 좋은 확산을 제공하기 때문에 일단 아무 보안 장치도 없다, 그래서 F는 비밀이 아니며 키가 필요하지도 않다. 여기서 섀넌은 좋은 혼합 함수를 확장해 다음과 같은 함수를 만든다면 혼돈도 추가할 수 있다고 덧붙인다.

$$ V_k(F(U_k(P_1, P_2, \cdots, P_n))) $$

여기에 나오는 $ U_k $와 $ V_k $는 키 값 k에 의해 결정되는, 단순 치환 암호처럼 비교적 복잡하지 않은 암호체계를 말한다. 함수 구조는 다음과 같아진다.<br>

위의 방식대로 설계된 함수는 처음부터 키를 바로 적용해서 혼합 함수에 의해서 확산시킨다. 즉, 키 정보가 모든 방향으로 분산되게 한다. (당연히 혼돈까지 추가된다.) 그리고 마지막으로 키 정보를 추가로 적용한다. 마지막 단계는 사실 혼돈을 제공하지 않는데, F가 공개되어 있기 때문에 F가 어떤 작업을 수행하더라도 즉시 무효화시킬 수 있다. 이러한 시도를 막기 위해 키 정보를 추가하는 단계가 반드시 필요한 것이다. 해석하자면, 혼합 함수의 작업을 무효화시켜서 혼합되었던 암호문을 혼합되지 않게 할 수 있다면, 진짜 $ U_k $라는 간단한 암호체계가 남게된다. 이 말은 다음과 상통한다. 다음과 같은 함수로 혼합을 진행하면 이는 보안 수주을 더 높일 수 있다.

$$ W_k(F_2(V_k(F_1(U_k(P_1, P_2, \cdots, P_n))))) $$

<strong>4.3. 치환 순열 네트워크</strong><br>
사실 군사적 목적에서 벗어나서 컴퓨터용 암호 체계를 본격적으로 연구가 시작된 것은 1970년 이후부터이다. 이를 연구하는 암호학자중에 호스트 페이스탈(Horst Feistel)이라는 사람은 섀넌의 암호시스템을 변형하여 현대에도 여전히 사용되고 있는 안전한 컴퓨터 암호를 생성하는 시스템을 두 가지 만들었다. <br>
첫 번째는 치환 순열 네트워크(Substitution-Permutation Network, SPN)인데, 이는 SP Network라고도 불린다.<br>
두 번째는 페이스탈 네트워크 (Feistel Network)이다.<br><br>

SP 네트워크는 일정한 패턴으로 치환 암호와 전치 암호를 사용한다. 섀넌의 혼합 함수 구축 방식처럼 큰 전치 암호 하나를 사용하고 힐 암호와 단일문자 치환 암호를 번갈아 사용하는 것이 아니라, 큰 전치 암호들과 작지만 복잡한 다중문자 치환 암호들을 번갈아 사용하고 덤으로 다표식 치환 암호 같은 것을 추가한다. 추가적으로 현대 암호 파트에 들어와서는 각 알파벳을 1 ~ 26의 숫자로 두고 연산하기 보다는, 이진수, 즉 비트를 사용해서 다룬다. 이에 대한 내용을 <strong>Overview/Data and operation In Binary Machine.ipynb</strong>에 서술해두었으니 확인하자. 

<img src="./src/pic_sp.png" width="300" height="400"><br>
<strong>Figure.4.3.1</strong><br>



본격적으로 네트워크의 구조를 살펴보자. 우선 현대 암호는 전형적으로 블록의 크기를 128비트로 두고있다. 평문이 128비트 있다고 일단 가정을 하겠다. 하지만, 키는 128비트이거나 더 클 수도 있다. 키 스케줄(Key schedule, 키들을 관리하는 시스템 또는 방안)에 따라 키를 부분 키로 분할하는데, 간단하게 앞에서부터 128비트씩 나눌 수도 있고 더 복잡한 방식을 써서 분할할 수도 있다. 어쨋든 최종적으로 각각 128비트인 일련의 라운드 키(round keys) $ K_0, K_1, K_2, \cdots $가 도출된다.<br><br>

 SP 네트워크에서 실제 암호화가 일어나는 첫 번째 단계는 <strong>치환</strong>이다. 이때, 키가 개입되는 방식은 평문의 비트와 첫 번째 라운드 키의 비트를 받아 올림 없이 moduler-2 summation에 대해서 다하고, 그 비트들을 여러 개의 작은 비트열로 나누는 것이다. 제안된 방식은 4비트씩 32개 비트열로 나누는 방식이다. 4비트로 구성된 비트열은 비기밀성 치환 박스인 S-박스를 통과하고, 이는 수학적으로 기술하기 복잡하게 만든 다중문자 치환 암호로서 4비트 단위로 실행된다. (당연히 32개 묶음이 돈다.) 이를 복잡하게 느끼는 일부 설계자들은 표를 하나 만들어서 그 표에 따라 모든 것을 처리하기도 한다고 한다. 이때 쓰이는 S-박스들은 모두 동일한 것일 수도 있고 서로 다른 것일 수도 있다. 중요한 것은 비기밀성 박스라는 사실이다. <br><br>

 두 번째로 암호화가 일어나는 단계는 <strong>순열</strong>이다. S-박스를 통과한 비트들은 다시 재결합되어 비기밀성 순열 박스인 P-박스를 통과하게 된다. 이때 P-박스는 128비트 전체 블록을 대상으로 전치를 수행한다. <br><br>

 자, 그러면 이제 언뜻 보기에도 확산이라는 성질이 굉장히 많이 추가되었다. 하지만, 여기서 끝이 아니다. 새로운 라운드 키를 받아서 또 치환과 순열연산을 진행하고 이러한 과정을 반복한다. 위의 그림에 나온 아키텍쳐는 어떤 방식인지 수업에서 확인하자.<br><br>

 섀넌의 혼합 함수 암호방식에 대해서 첫 번째와 마지막 단계에서 키가 설정되어야 한다. 만약에 그렇지 않으면 사실상 비기밀성 연산과정인 S-박스와 P-박스에서 복호화를 그냥 진행하면된다. 결론적으로는 각 4비트 단위당 확산과 혼돈을 추가하기 위해 치환을 하고, 이의 결과에 대해서 전체적으로(128비트) 확산시키기 위해서 순열연산을 진행한다. 물론 복호화는 라운드키를 이용하여 각 단계를 거꾸로 수행하면 된다. 결국 확산에 의한 쇄도효과(한 비트만 변해도 출력값 비트 중 상당 부분이 변하는 성질)이 잘 이루어 졌다고 할 수 있다.<br><br>

 현대 암호학자들은 이 쇄도효과에 대해서 엄밀하게 쇄도 기준을 만들었다.(쉽게 말하면, 한 비트만 변했을 때, 얼마나 많은 출력값의 비트가 변해야하는지에 대한 기준) 다음과 같이 크기가 작은 3비트짜리 예를 살펴보자.

위의 표는 가운데 비트가 0인 데이터가 S-박스를 통과했을 때, 출력값의 마지막 비트가 어떻게 되는지 살펴보면, 마지막 비트가 바뀐 경우는 두 건이고, 바뀌지 않은 것이 두 건이다. 입력값의 어느 비트를 선택하고 출력값의 어느 비트를 선택하든 언제나 같은 현상이 일어난다는 것을 확인할 수 있다. 이러한 배경아래에서 엄밀한 쇄도 기준은 다음과 같이 정의된다.

> 만약 입력값 중 한 비트만 바뀌고 나머지는 그대로라면, 그대로인 비트에 대응하는 출력값 가운데 절반은 바뀌고 절반은 바뀌지 않는 것.

즉, 잘 분산되어진다.<br><br>

자, 이렇게 엄밀한 쇄도 기준을 만족하면서 수학적으로 복잡한 128비트 S-박스를 만들 수 있다면 확산과 혼돈의 성질을 모두 만족한다. 그런데 당연히 이런 암호를 구현할 수 없고, 심지어 현대 기술로도 어렵다. (굉장히 이상적인 암호화다.) 그래서 일반적으로 사용되는 SP 네트워크에서는 일단 S-박스를 통과해서 값이 바뀐 비트가 여러개 나오면 P-박스를 사용해 그 비트들을 전체 블록으로 골고루 분산시킨다. P-박스를 통과한 비트들이 다시 S-박스를 통과하면 더 많은 비트가 바뀌고, 이떄, 모든 S-박스에서 엄밀한 쇄도 기준이 충족되었고, P-박스가 주도면밀하게 구성되었으며 충분하게 여러 차례의 라운드가 사용된다면 결국 입력되는 평문에서 비트 하나가 변했을 때 출력되는 암호문 비트가 모두 변할 수 있는 확률이 50%가 되도록 할 수 있다.

이는 SP 네트워크처럼 키 스케줄을 사용해서 라운드 키를 생성한다. 입력되는 평문의 비트를 반씩 두 부분으로 나눈다. 절반 비트가 라운드 함수(round function)f를 거친다. 라운드 함수는 일반적으로 하나 이상의 S-박스와 P-박스, 라운드 키 비트의 moduler-2 summation으로 구성된다. (SP 네트워크의 한 개의 계층과 유사하다) 라운드 함수를 통과해서 나온 비트를 남아있던 왼쪽 절반의 입력값 비트에 moduler-2 summation 연산을 수행한다. 한 라운드가 끝나면 오른쪽 절반과 왼쪽 절반 비트의 자리를 바꿔 앞선 과정을 반복한다. 보통 10~20회를 반복하는 것이 일반적이다. 마지막 라운드가 끝나면 자리를 바꾸지 않는다. 그렇더라도 암호의 안정성에 영향을 미치지 않으며 오히려 복호화할 떄 더 편리하다. 이 친구는 복호화를 할 때 암호화를 역으로 실행하는 것이 아니라 키 스케줄을 제외하고는 암호화와 같은 방향으로 진행된다. (moduler-2 summation과 moduler-2 subtract가 같기 때문이다.) 덕분에 S-박스가 거꾸로 수행할 필요가 없기 때문에 쇄도 효과와 같은 좋은 성질을 갖는 S-박스를 선택할 수 있는 여지가 더욱 많다. (이떄, S-박스는 입력값의 비트 수와 출력값의 비트 수를 다르게 할 수 있다. P-박스도 본래의 순열뿐만 아니라, 확장 함수나 압축 함수를 사용할 수 있다.) <br><br>

SP 네트워크와 마찬가지로 파이스텔 네트워크에서도 라운드 함수에 있는 S-박스가 혼돈을 제공한다. 그런 다음 P-박스에서 전체적으로 확산을 제공하고, moduler-2 summation을 거치며 확산과 혼돈이 나머지 절반으로 전파된다. <br><br>

페이스탈 네트워크 암호 중에서 가장 유명한 것은 암호화 표준(DES)이다. 여기에 대해서는 다음시간에 살펴보도록 하자.
